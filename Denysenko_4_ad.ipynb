# Використання Pandas

```python
import pandas as pd
import numpy as np
import timeit
from datetime import datetime
#Зчитаємо файл
df = pd.read_csv('C:\\Users\\Anastasiia\\Desktop\\household_power_consumption.txt', sep = ';',  decimal='/')
df.head()
```

```python
start_time = 0.0
def start_count():
    global start_time
    start_time = time.time()
def finish_count():
    print(time.time()-start_time)
```

```python
#Видалимо строки де немає даних
data1=df[ ~df.Voltage.str.contains("\?") ]
data = pd.DataFrame(data1)
#Перевіримо чи данні видалені
no_data = data.loc[df['Global_intensity'] == '?']
print(no_data)
```

```python
#Змінимо типи даних
data['Global_active_power'] = data['Global_active_power'].astype(float)
data['Global_reactive_power'] = data['Global_reactive_power'].astype(float)
data['Voltage'] = data['Voltage'].astype(float)
data['Global_intensity'] = data['Global_intensity'].astype(float)
data['Sub_metering_1'] = data['Sub_metering_1'].astype(float)
data['Sub_metering_2'] = data['Sub_metering_2'].astype(float)
data['Sub_metering_3'] = data['Sub_metering_3'].astype(float)
```

```python
#Знайдемо господарства, де загальна активна споживана потужність перевищує 5 кВт.
start_count()
greater_than_5 = data.loc[(data['Global_active_power'] > 5)]
finish_count()
greater_than_5.head()
```

```python
#Знайдемо господарства, де вольтаж перевищує 235 Вольт.
start_count()
greater_than_235 = data.loc[(data['Voltage'] > 235)]
finish_count()
greater_than_235.head()
```

```python
#Знайдемо господарства, де сила струму в діапазоні 19-20 Ампер.
start_count()
in_19_20 = data.loc[(data['Global_intensity']>19)&(data['Global_intensity']<20)&(data['Sub_metering_2']>data['Sub_metering_3'])]
finish_count()
in_19_20.head()
```

```python
#Виберемо 50000 випадкових значень
start_count()
random_50000 = data.sample(n=50000)
finish_count()
random_50000.head()
```

```python
#Знайдемо середнє значення
start_count()
print(random_50000['Sub_metering_1'].mean())
print(random_50000['Sub_metering_2'].mean())
print(random_50000['Sub_metering_3'].mean())
print("***")
finish_count()
```

```python
#Знайдемо домогосподарства, де після 18:00 Sub_metering_2 має більше знначення 
start_count()
later_than_18 = data.loc[(data['Time'] > '18:00:00')&(data['Sub_metering_2']>data['Sub_metering_3'])&(data['Sub_metering_2']>data['Sub_metering_1'])]
#Розіб'ємо на половини та виведемо для першої половини кожен третій результат, я для другої - кожний четвертий.
first_half = (later_than_18.iloc[:38825,:]).iloc[::3, :]
second_half = (later_than_18.iloc[38825:,:]).iloc[::4, :]
#Поєднаємо їх
result = pd.concat([first_half, second_half])
finish_count()
result.head()
```

# Використання Numpy

```python
#Прочитаємо дані як масив numpy
dn = data.to_numpy()
#Знайдемо господарства, де загальна активна споживана потужність перевищує 5 кВт.
start_count()
d1 = dn[np.where(dn[...,2]>5)]
finish_count()
print(d1)
```

```python
#Знайдемо господарства, де вольтаж перевищує 235 Вольт.
start_count()
d2 = dn[np.where(dn[...,4]>235)]
finish_count()
print(d2)
```

```python
#Знайдемо господарства, де сила струму в діапазоні 19-20 Ампер.
start_count()
d3 = dn[np.where((dn[...,5]>19)&(dn[...,5]<20)&(dn[...,7]>dn[...,8]))]
finish_count()
print(d3)
```

```python
#Виберемо 50000 випадкових значень
start_count()
print(dn[np.random.choice(dn.shape[0], 50000, replace = False)])
finish_count()
```

```python
#Знайдемо середнє значення
start_count()
print(np.mean(dn[...,6]))
print(np.mean(dn[...,7]))
print(np.mean(dn[...,8]))
print("***")
finish_count()
```

```python
#Розіб'ємо на половини та виведемо для першої половини кожен третій результат, я для другої - кожний четвертий.
start_count()
d6 = dn[np.where((dn[...,1]>'18:00:00')&(dn[...,2]>6)&(dn[...,7]>dn[...,6])&(dn[...,7]>dn[...,8]))]
new = np.split(d6, [529, 530])
print(new[0][::3])
print(new[1][::4])
finish_count()
```

I. Пошук домогосподарств, де загальна активна споживана потужність перевищує 5 кВт, зайняв 0.004009723663330078 с в pandas та 0.03000330924987793 с в numpy, отже, з цю задачу pandas виконав швидше.
II. Пошук домогосподарств, де вольтаж перевищує 235 Вольт, зайняв 0.0616917610168457 с в pandas та 0.0807197093963623 с в numpy, отже, з цю задачу pandas виконав швидше.
III. Пошук домогосподарств, де сила струму в діапазоні 19-20 Ампер та Sub_metering_2 > Sub_metering_3, зайняв 0.007335662841796875 с в pandas та 0.07647871971130371 с в numpy, отже, з цю задачу pandas виконав трохи швидше.
IV. Знаходження середнього значення в pandas зайняло 0.0010006427764892578 с, а в numpy 0.04952836036682129, отже, з цю задачу pandas теж виконав швидше.
V. Знаходження домогосподарств, де після 18:00 Sub_metering_2 має більше знначення та виведення кожного третього та кожного четвертого зачення в першій та другій половинах відповідно зайняло 0.05932760238647461 с в pandas та 0.08147811889648438 с в numpy, отже, pandas виконав цю задачу швидше.
