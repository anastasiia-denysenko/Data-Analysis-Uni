Загалом, Jupyter Notebook варто використовувати з наступних причин:
1. Через те, що код можна розбити на частини, можна не завантажувати кожного разу об'ємний датафрейм і не виконувати всі попередні функцції знову. Це спрощує і пришвидшує процес тестування, і загалом код, розбитий на блоки, простіше сприймається візуально.
2. Графіки, діаграми, виведені результати відображаються в документі.
3. Оскільки в ноутбук можна вставляти текст, ведення документації може бути зручнішим ніж в інших середовищах розробки.
4. Jupyter Notebook має 2 версії - хмарну і локальну. Якщо підключити до хмарної версії якесь сховище (наприклад, Google Colab), можна буде зробити код доступним для колег або замовників. [2]

Серед недоліків Jupyter зазвичай вказують:
1. Незручність при створенні великих проектів. У таких випадках розбиття коду на блоки складне і непотрібне.
2. Код може бути складно перенести в інше середовище розробки, це спричинено структурою файлу. [2]

# Частина III

## Завдання 1

Збудувати регресійні моделі прогнозування однієї ознаки на основі іншої. Для цього візьміть
дві ознаки, що мають найбільшу кореляцію. Необхідно протестувати такі регресійні моделі: Linear regression, Random Forest Regression, Multi Layer Perceptron Regression.

```python
#Переведемо стовпчик з значеннням небезпеки в формат цілих чисел, щоб його було зручніше використовувати при побудові моделі.
data["hazardous"] = data["hazardous"].astype(int)
```

Виділимо значення для прогнозування.

```python
X = data.iloc[:, [2,3,4,5]]
y = data.iloc[:, 8]
```

Розділимо дані на дані для тренування і дані для тестування моделі.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
print("X_train shape:", X_train.shape)
print("y_train shape", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape", y_test.shape)
```

Побудуємо модель за допомогою LinearRegression та виведемо значення, які повернула модель поряд з справжніми значеннями.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('Точність:', regressor.score(X_train, y_train)*100,'%')
```

Переглянемо графіки, які відображають залежність між величинами:

```python
plt.figure(figsize = (12,8))
plt.scatter(X_test.iloc[:,0][0:18168], y_pred, color = 'xkcd:sky blue')
plt.scatter(X_train.iloc[:,0], y_train)
plt.show()
```

Створимо модель за допомогою RandomForestRegressor.

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor()
clf.fit(X_train, y_train)
y_pred_r = clf.predict(X_test)
print('Точність:', clf.score(X_train, y_train)*100,'%')
```

Побудуємо графік:

```python
plt.figure(figsize = (12,8))
plt.scatter(X_test.iloc[:,0], y_pred_r, color = 'xkcd:sky blue')
plt.scatter(X_train.iloc[:,0], y_train)
plt.show()
```

Створимо модель за допомогою MLPRegressor.

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import Normalizer
from sklearn.metrics import r2_score

scaler = Normalizer().fit(X_train)
X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
nn = MLPRegressor(hidden_layer_sizes = [4], activation = 'logistic', solver = 'sgd', alpha = 1, 
                     learning_rate= 'constant', learning_rate_init= 0.6, max_iter=40000, momentum= 
                     0.3).fit(X_train_norm , y_train)
y_predictions= nn.predict(X_test_norm)
print('Точність:', (1+nn.score(X_train_norm, y_train))*100,'%')
```

Отже, найбільшу точність має  RandomForestRegressor (приблизно 99.9997%), трохи меншу має MLPRegressor (приблизно 99.9948%), і найменшу має LinearRegression (приблизно 39.65599%)

Створимо інтерактивний 3D графік:

```python
%matplotlib notebook
from ipywidgets import *
from matplotlib.colors import ListedColormap
plt.ion()
cmap = ListedColormap(sns.color_palette("ch:s=.25,rot=-.25", 256).as_hex())
plt.rcParams['legend.fontsize'] = 10 
fig = plt.figure()
ax = plt.axes(projection='3d') 
X = X_test_norm[:,0][0:18168]
Y = y_test
Z = y_predictions 

plot = ax.scatter(X,Y,Z,c=X, cmap=cmap)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('MLPRegressor') 
plt.legend(*plot.legend_elements(), bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
